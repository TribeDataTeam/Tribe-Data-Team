```python
# removing unwanted columns

com_reg_ext.drop(
    columns=[
        "name",
        "Email Date",
        "Do you have a startup/business idea that you want to work on?",
    ],
    inplace=True,
)

com_reg_ext
```

```python
# next we rename columns

com_reg_ext.rename(
    columns={
        "What part of the community circle are you most interested in?": "Interested part of the com",
        "What skills are you looking forward to developing during your time in the Community Circle": "Skills to develop",
        "What attributes are you aiming to get through the Community Circle?": "Expectations",
        "Where or from whom did you hear about Fellowship?": "Source of Info",
    },
    inplace=True,
)
```

```python
com_reg_ext
```

```python
# Next we use the explode function to separate columns with multiple values

pref_int = (
    com_reg_ext["Interested part of the com"]
    .str.split(",")
    .explode()
    .reset_index(name="Interest in comm.")
)
pref_skill = (
    com_reg_ext["Skills to develop"]
    .str.split(",")
    .explode()
    .reset_index(name="Skill to develop")
)
pref_expectations = (
    com_reg_ext["Expectations"]
    .str.split(",")
    .explode()
    .reset_index(name="Expectations")
)

# Now we merge the ouput of these exploded columns

merged_data = pd.merge(
    pref_int, pref_skill, left_on="index", right_on="index", how="outer"
).merge(pref_expectations, left_on="index", right_on="index", how="outer")

merged_data_02 = pd.merge(com_reg_ext, merged_data, left_index=True, right_on="index")

merged_data_02.columns

community_reg_data = merged_data_02[
    [
        "index",
        "id",
        "Name",
        "Status",
        "email",
        "Date of Application",
        "Interested part of the com",
        "Interest in comm.",
        "Skills to develop",
        "Skill to develop",
        "Expectations_x",
        "Expectations_y",
        "Source of Info",
        "Date of onboarding",
        "Do you live in Tampere area?",
    ]
]

community_reg_data
```

```python
from fuzzywuzzy import process


# 1. Preprocessing Function =================================================
def preprocess_sources(text_series):
    # Define typo corrections and term standardization
    typo_map = {
        r"\blinkedin\b": "linkedin",
        r"\blinked in\b": "linkedin",
        r"\blinkdein\b": "linkedin",
        r"\bgoogle search\b": "google",
        r"\bsearch engine\b": "google",
        r"\bte office\b": "te office",
        r"\binternational house\b": "international house tampere",
        r"\btampere international house\b": "international house tampere",
        r"\btamepere international house\b": "international house tampere",
        r"\btribe website\b": "tribe website",
        r"\binternet site\b": "website",
        r"\bonline\b": "website",
        r"\bplatform 6\b": "platform6",
        r"\bp6\b": "platform6",
    }

    # Apply preprocessing
    cleaned = (
        text_series.str.lower().str.strip().fillna("")  # Handle NaN values
    )

    # Apply regex replacements sequentially
    for pattern, replacement in typo_map.items():
        cleaned = cleaned.str.replace(pattern, replacement, regex=True)

    return cleaned


# Apply preprocessing
community_reg_data["cleaned_source"] = preprocess_sources(
    community_reg_data["Source of Info"]
)

# 2. Enhanced Category Mapping ===============================================
category_map = {
    "digital channels": [
        "website",
        "internet",
        "online",
        "google",
        "search",
        "linkedin",
        "social media",
        "instagram",
        "reddit",
    ],
    "events": [
        "event",
        "job fair",
        "networking",
        "after work",
        "tribe event",
        "tribe activities",
        "hubs event",
    ],
    "organizations": [
        "te office",
        "tamk",
        "international house tampere",
        "business tampere",
        "house of tampere",
        "proakatemia",
        "tribe venue",
        "platform6",
        "ensimetri",
        "chips",
    ],
    "personal referrals": [
        "friend",
        "colleague",
        "word of mouth",
        "classmate",
        "school friend",
        "sister",
        "entrepreneur",
        "told me",
    ],
    "internal connections": [
        "intern",
        "fellowship",
        "board",
        "operational team",
        "community development",
        "part of",
        "tribe community",
    ],
    "specific people": [
        "julia",
        "katerina",
        "santeri",
        "jarmo",
        "jubril",
        "saadman",
        "pranav",
        "michelle",
        "jaakko",
        "victor",
        "abdel",
        "ishan",
        "analyn",
        "kat",
    ],
}


# 3. Hybrid Categorization Function ==========================================
def apply_category(source_text):
    if not source_text:
        return "uncategorized"

    # First pass: term matching
    for category, terms in category_map.items():
        if any(term in source_text for term in terms):
            return category

    # Second pass: fuzzy matching
    all_terms = [term for sublist in category_map.values() for term in sublist]
    matches = process.extract(source_text, all_terms, limit=1)

    if matches and matches[0][1] >= 75:
        matched_term = matches[0][0]
        for category, terms in category_map.items():
            if matched_term in terms:
                return category

    return "uncategorized"


# Apply categorization
community_reg_data["category"] = community_reg_data["cleaned_source"].apply(
    apply_category
)


# 4. Post-Processing and Final Analysis ======================================
# For items still uncategorized, use a secondary approach
def refine_uncategorized(row):
    if row["category"] != "uncategorized":
        return row["category"]

    source = row["cleaned_source"]

    # Look for specific patterns
    if "tribe" in source:
        return "internal connections"
    elif any(name in source for name in ["julia", "jaakko", "michelle", "kat"]):
        return "specific people"
    elif any(term in source for term in ["friend", "told", "heard"]):
        return "personal referrals"
    elif any(term in source for term in ["event", "networking"]):
        return "events"

    return "uncategorized"


community_reg_data["final_category"] = community_reg_data.apply(
    refine_uncategorized, axis=1
)

community_reg_data.columns

# replacing values

community_reg_data["Do you live in Tampere area?"].replace("v", "Yes", inplace=True)
```
