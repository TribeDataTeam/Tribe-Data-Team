# Pattern identification and Grouping for preferred topic column

```python
# We shall use fuzzywuzzy to automatically categorize similar responses into same groups

#first we import our library

from fuzzywuzzy import process

```

```python
#Now we extract columns required for our pattern identification operation
req_data = cleaned_data[['id', 'index', 'Event Date', 'Pref_days', 'pref_times', 'pref_topic' ]]
```

```python
# 1. Enhanced Preprocessing ==================================================
def preprocess_topics(text_series):
    # This aspect handles typos in the text found in the column of interest. Here we define typo corrections
    typo_map = {
        r'\bbeginig\b': 'beginning',
        r'\bentrepeneur\b': 'entrepreneur',
        r'\bstart ip\b': 'startup',
        r'\bsparring\b': 'brainstorming',
        r'\bnil\b': '',
        r'\.': ''
    }
    
    # Apply preprocessing
    cleaned = (
        text_series
        .str.lower()
        .str.strip()
    )
    
    # Apply regex replacements sequentially
    for pattern, replacement in typo_map.items():
        cleaned = cleaned.str.replace(pattern, replacement, regex=True)
    
    # Final cleanup
    cleaned = (
        cleaned
        .str.replace(r'[^\w\s-]', '', regex=True)
        .str.replace(r'\s+', ' ', regex=True)
    )
    
    return cleaned

# Apply preprocessing
cleaned_data['pref_topic'] = preprocess_topics(cleaned_data['pref_topic'])

# 2. Enhanced Clustering =====================================================
def cluster_topics(topic_list, threshold=80):
    clusters = {}
    sorted_topics = sorted(topic_list, key=len, reverse=True)
    
    for topic in sorted_topics:
        matches = process.extract(topic, clusters.keys(), limit=1)
        
        if matches and matches[0][1] >= threshold:
            base_term = matches[0][0]
            if any(word in topic for word in base_term.split()):
                clusters[base_term].append(topic)
                continue
                
        clusters[topic] = [topic]
    
    return clusters

clusters = cluster_topics(cleaned_data['pref_topic'].tolist(), threshold=85)

# 3. Enhanced Category Mapping ===============================================
category_map = {
    'startup & entrepreneurship': [
        'startup', 'entrepreneur', 'founder', 'venture',
        'lessons learned', 'success stories', 'how to start',
        'employer regulations', 'startup life'
    ],
    'tech': [
        'tech', 'technology', 'ict', 'it', 'ai', 'software',
        'web3', 'technical skills', 'machine learning'
    ],
    'business': [
        'business', 'business development', 'business ideas',
        'business theories', 'business models', 'new business'
    ],
    'marketing & sales': [
        'marketing', 'sales', 'customer acquisition',
        'brand building', 'growth hacking'
    ],
    'finance & legal': [
        'finance', 'funding', 'invoicing', 'legal',
        'investment', 'taxation'
    ],
    'networking': [
        'networking', 'connection', 'collaboration',
        'knowledge exchange', 'community building'
    ],
    'ux & design': [
        'ux design', 'user experience', 'psychology',
        'interface design', 'user research'
    ],
    'industry specific': [
        'beauty industry', 'health tech', 'fintech',
        'e-commerce', 'saas'
    ]
}

# 4. Hybrid Categorization ===================================================
def apply_category(topic):
    # First pass: exact matching
    for key, terms in category_map.items():
        if any(term in topic for term in terms):
            return key
            
    # Second pass: fuzzy term matching
    all_terms = [term for sublist in category_map.values() for term in sublist]
    matches = process.extract(topic, all_terms, limit=1)
    
    if matches and matches[0][1] >= 75:
        matched_term = matches[0][0]
        for key, terms in category_map.items():
            if matched_term in terms:
                return key
                
    # Third pass: cluster-based matching
    for cluster_base, cluster_topics in clusters.items():
        if topic in cluster_topics:
            cluster_matches = process.extract(cluster_base, category_map.keys(), limit=1)
            if cluster_matches and cluster_matches[0][1] >= 70:
                return cluster_matches[0][0]
    
    return 'uncategorized'

cleaned_data['category'] = cleaned_data['pref_topic'].apply(apply_category)

# 5. Post-Processing =========================================================
category_merge_rules = {
    'tech': ['ai', 'web3'],
    'startup & entrepreneurship': ['founder stories'],
    'finance & legal': ['taxation']
}

cleaned_data['final_category'] = cleaned_data['category'].replace({
    'industry specific': 'specialized',
    'ux & design': 'ux'
})

# 6. Analysis & Export =======================================================
print(cleaned_data['final_category'].value_counts())

# Export remaining uncategorized topics
#uncategorized = req_data[req_data['final_category'] == 'uncategorized']
#uncategorized.to_csv('uncategorized_topics.csv', index=False)

# Export categorized results
#req_data.to_csv('categorized_topics.csv', index=False)
```
